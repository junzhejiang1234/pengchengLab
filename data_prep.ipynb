```python
def add_tech_ind(df):
    df = df.copy()
    def ma(arr, n):
        if n == 1:
            return arr
        else:
            result = [0 for i in range(n - 1)]
            result1 = arr[:-n + 1].copy()
            length = len(arr)
            for i in range(1, n):
                result1 += arr[i:length + 1 - n + i]
            result1 /= n
            result += list(result1)
            return np.array(result)
    
    def ema(arr, n):
        if n == 1:
            return arr
        else:
            result = []
            for i in range(len(arr)):
                temp = arr[i]
                if i == 0:
                    result.append(temp)
                else:
                    temp = 2 / (1 + n) * temp
                    temp += (1 - 2 / (1 + n)) * result[-1]
                    result.append(temp)
            return np.array(result)
        
    def prev_highs(arr, n):
        if n == 1:
            return arr
        else:
            length = len(arr)
            result = [np.nanmax(arr[max(i - n + 1, 0): i + 1]) for i in range(length)]
            return np.array(result)
    
    def prev_lows(arr, n):
        if n == 1:
            return arr
        else:
            length = len(arr)
            result = [np.nanmin(arr[max(i - n + 1, 0): i + 1]) for i in range(length)]
            return np.array(result)
    
    def stds(arr, n):
        if n == 1:
            print('False!!!')
        else:
            length = len(arr)
            result = [np.nanstd(arr[max(i - n + 1, 0): i + 1]) for i in range(1, length)]
            result = [0] + result
            return np.array(result)
    
    def add_ma_ret(df):
        ''' add ma of return calculated by closing price '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ret = cps[1:] / cps[:-1] - 1.0
        ret1 = [0.0] + list(ret)
        df['ret1'] = ret1
        ret1  = np.array(ret1)
        ret2 = [0.0] + list((ret1[1:] + ret1[:-1]) / 2)
        df['ret2'] = ma(ret1, 2)
        df['ret4'] = ma(ret1, 4)
        df['ret8'] = ma(ret1, 8)
        df['ret16'] = ma(ret1, 16)
        df['ret32'] = ma(ret1, 32)
        df['ret64'] = ma(ret1, 64)
        df['ret128'] = ma(ret1, 128)
        
    def add_ema_ret(df):
        ''' add ma of return calculated by closing price '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ret = cps[1:] / cps[:-1] - 1.0
        ret1 = [0.0] + list(ret)
#         df['ret1'] = ret1
        ret1  = np.array(ret1)
#         ret2 = [0.0] + list((ret1[1:] + ret1[:-1]) / 2)
        df['ret2e'] = ema(ret1, 2)
        df['ret4e'] = ema(ret1, 4)
        df['ret8e'] = ema(ret1, 8)
        df['ret16e'] = ema(ret1, 16)
        df['ret32e'] = ema(ret1, 32)
        df['ret64e'] = ema(ret1, 64)
        df['ret128e'] = ema(ret1, 128)
        
    def add_std_ret(df):
        ''' add std of return calculated by closing price '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ret = cps[1:] / cps[:-1] - 1.0
        ret1 = [0.0] + list(ret)
        ret1  = np.array(ret1)
        df['std2'] = stds(ret1, 2)
        df['std4'] = stds(ret1, 4)
        df['std8'] = stds(ret1, 8)
        df['std16'] = stds(ret1, 16)
        df['std32'] = stds(ret1, 32)
        df['std64'] = stds(ret1, 64)
        df['std128'] = stds(ret1, 128)
        
        
    def add_ma_vol(df):
        ''' add ma of volume '''
        vol = df.loc[:, 'Volume'].values
        vol = np.array([float(val) for val in vol], dtype = 'float')
        df['vol2'] = ma(vol, 2)
        df['vol4'] = ma(vol, 4)
        df['vol8'] = ma(vol, 8)
        df['vol16'] = ma(vol, 16)
        df['vol32'] = ma(vol, 32)
        df['vol64'] = ma(vol, 64)
        df['vol128'] = ma(vol, 128)
        
    def add_vwap_c(df):
        ''' add difference between vwap and close'''
        df['vcb1'] = df['VWAP'].values / df['Close'].values - 1
        df['vcb2'] = ma(df['vcb1'].values, 2)
        df['vcb4'] = ma(df['vcb1'].values, 4)
        df['vcb8'] = ma(df['vcb1'].values, 8)
        df['vcb16'] = ma(df['vcb1'].values, 16)
        df['vcb32'] = ma(df['vcb1'].values, 32)
        df['vcb64'] = ma(df['vcb1'].values, 64)
        df['vcb128'] = ma(df['vcb1'].values, 128)
        
    def add_weekday_dummyvar(df):
        ''' add what week day it is '''
        dates = df['Date'].values
        dates = [datetime.date(int(val[:4]), int(val[5:7]), int(val[8:])) for val in dates]
        weekday = [(val.weekday() == 0 or val.weekday() == 4)  for val in dates]
        result = np.zeros(len(df))
        result[weekday] = 1
        df['weekday'] = result
        
    def add_subline(df):
        ''' add k-line subline '''
        length = len(df)
        result = np.ones(length)
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        result = np.min(np.vstack((ops, cps)), axis = 0)
        df['sub'] = (result - lows) / cps
        ops_2 = np.array([ops[0]] * 1 + list(ops[:-1]))
        df['sub2'] = (np.min(np.vstack((ops_2, cps)), axis = 0) - prev_lows(lows, 2)) / cps
        ops_4 = np.array([ops[0]] * 3 + list(ops[:-3]))
        df['sub4'] = (np.min(np.vstack((ops_4, cps)), axis = 0) - prev_lows(lows, 4)) / cps
        ops_8 = np.array([ops[0]] * 7 + list(ops[:-7]))
        df['sub8'] = (np.min(np.vstack((ops_8, cps)), axis = 0) - prev_lows(lows, 8)) / cps
        ops_16 = np.array([ops[0]] * 15 + list(ops[:-15]))
        df['sub16'] = (np.min(np.vstack((ops_16, cps)), axis = 0) - prev_lows(lows, 16)) / cps
        ops_32 = np.array([ops[0]] * 31 + list(ops[:-31]))
        df['sub32'] = (np.min(np.vstack((ops_32, cps)), axis = 0) - prev_lows(lows, 32)) / cps
        ops_64 = np.array([ops[0]] * 63 + list(ops[:-63]))
        df['sub64'] = (np.min(np.vstack((ops_64, cps)), axis = 0) - prev_lows(lows, 64)) / cps
        ops_128 = np.array([ops[0]] * 127 + list(ops[:-127]))
        df['sub128'] = (np.min(np.vstack((ops_128, cps)), axis = 0) - prev_lows(lows, 128)) / cps
        
    def add_macd(df):
        ''' add MACD '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        df['MACD'] = (ema(cps, 12) - ema(cps, 26)) / cps
        
    def add_bbbb_day(df):
        ''' add bbbb in day '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        highs = df.loc[:,'High'].values
        highs = np.array([float(val) for val in highs])
        df['bbbb_1'] = ((highs + lows) - 2.0 * cps) / cps
        df['bbbb_2'] = ((prev_highs(highs, 2) + prev_lows(lows, 2)) - 2.0 * cps) / cps
        df['bbbb_4'] = ((prev_highs(highs, 4) + prev_lows(lows, 4)) - 2.0 * cps) / cps
        df['bbbb_8'] = ((prev_highs(highs, 8) + prev_lows(lows, 8)) - 2.0 * cps) / cps
        df['bbbb_16'] = ((prev_highs(highs, 16) + prev_lows(lows, 16)) - 2.0 * cps) / cps
        df['bbbb_32'] = ((prev_highs(highs, 32) + prev_lows(lows, 32)) - 2.0 * cps) / cps
        df['bbbb_64'] = ((prev_highs(highs, 64) + prev_lows(lows, 64)) - 2.0 * cps) / cps
        df['bbbb_128'] = ((prev_highs(highs, 128) + prev_lows(lows, 128)) - 2.0 * cps) / cps
    
    def add_cccc_day(df):
        ''' add cccc in day '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        highs = df.loc[:,'High'].values
        highs = np.array([float(val) for val in highs])
        df['cccc_1'] = (highs - lows) / 2 / cps
        df['cccc_2'] = (prev_highs(highs, 2) - prev_lows(lows, 2)) / 2.0 / cps
        df['cccc_4'] = (prev_highs(highs, 4) - prev_lows(lows, 4)) / 2.0 / cps
        df['cccc_8'] = (prev_highs(highs, 8) - prev_lows(lows, 8)) / 2.0 / cps
        df['cccc_16'] = (prev_highs(highs, 16) - prev_lows(lows, 16)) / 2.0 / cps
        df['cccc_32'] = (prev_highs(highs, 32) - prev_lows(lows, 32)) / 2.0 / cps
        df['cccc_64'] = (prev_highs(highs, 64) - prev_lows(lows, 64)) / 2.0 / cps
        df['cccc_128'] = (prev_highs(highs, 128) - prev_lows(lows, 128)) / 2.0 / cps
        
    def add_kdj(df):
        ''' add kdj '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        highs = df.loc[:,'High'].values
        highs = np.array([float(val) for val in highs])
        
        # high freq
        K = np.zeros(len(df))
        for i in range(len(df)):
            start = max(i - 4, 0)
            end = i + 1
            K[i] = (cps[i] - np.min(lows[start:end])) / (np.max(highs[start:end]) - np.min(lows[start:end]))
        D = ema(K, 3)
        df['K'] = K
        df['D'] = D
        temp = np.ones(len(df))
        temp[K > 0.7] = 0
        temp[K < 0.3] = 0
        temp[K < D] = 0
        df['KDJ'] = temp
        
        # low freq shit
        K = np.zeros(len(df))
        for i in range(len(df)):
            start = max(i - 13, 0)
            end = i + 1
            K[i] = (cps[i] - np.min(lows[start:end])) / (np.max(highs[start:end]) - np.min(lows[start:end]))
        D = ema(K, 3)
        df['K_'] = K
        df['D_'] = D
        temp = np.ones(len(df))
        temp[K > 0.7] = 0
        temp[K < 0.3] = 0
        temp[K < D] = 0
        df['KDJ_'] = temp
        
    def standardize_ohlv(df):
        ''' divide open high low by vwapclose'''
        df['Open'] = df['Open'].values / df['Close'].values
        df['High'] = df['High'].values / df['Close'].values
        df['Low'] = df['Low'].values / df['Close'].values
        df['VWAP'] = df['VWAP'].values / df['Close'].values
        
    
    add_ma_ret(df)
    add_ema_ret(df)
    add_std_ret(df)
    add_ma_vol(df)
    add_weekday_dummyvar(df)
    add_subline(df)
    add_macd(df)
    add_bbbb_day(df)
    add_cccc_day(df)
    add_kdj(df)
    standardize_ohlv(df)
    return df
```

```python
def add_tech_ind_btc(df):
    df = df.copy()
    def ma(arr, n):
        if n == 1:
            return arr
        else:
            result = [0 for i in range(n - 1)]
            result1 = arr[:-n + 1].copy()
            length = len(arr)
            for i in range(1, n):
                result1 += arr[i:length + 1 - n + i]
            result1 /= n
            result += list(result1)
            return np.array(result)
    
    def ema(arr, n):
        if n == 1:
            return arr
        else:
            result = []
            for i in range(len(arr)):
                temp = arr[i]
                if i == 0:
                    result.append(temp)
                else:
                    temp = 2 / (1 + n) * temp
                    temp += (1 - 2 / (1 + n)) * result[-1]
                    result.append(temp)
            return np.array(result)
        
    def prev_highs(arr, n):
        if n == 1:
            return arr
        else:
            length = len(arr)
            result = [np.nanmax(arr[max(i - n + 1, 0): i + 1]) for i in range(length)]
            return np.array(result)
    
    def prev_lows(arr, n):
        if n == 1:
            return arr
        else:
            length = len(arr)
            result = [np.nanmin(arr[max(i - n + 1, 0): i + 1]) for i in range(length)]
            return np.array(result)
    
    def stds(arr, n):
        if n == 1:
            print('False!!!')
        else:
            length = len(arr)
            result = [np.nanstd(arr[max(i - n + 1, 0): i + 1]) for i in range(1, length)]
            result = [0] + result
            return np.array(result)
    
    def add_ma_ret(df):
        ''' add ma of return calculated by closing price '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ret = cps[1:] / cps[:-1] - 1.0
        ret1 = [0.0] + list(ret)
        df['ret1'] = ret1
        ret1  = np.array(ret1)
        ret2 = [0.0] + list((ret1[1:] + ret1[:-1]) / 2)
        df['ret2'] = ma(ret1, 2)
        df['ret4'] = ma(ret1, 4)
        df['ret8'] = ma(ret1, 8)
        df['ret16'] = ma(ret1, 16)
        df['ret32'] = ma(ret1, 32)
        df['ret64'] = ma(ret1, 64)
        df['ret128'] = ma(ret1, 128)
        
    def add_ema_ret(df):
        ''' add ma of return calculated by closing price '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ret = cps[1:] / cps[:-1] - 1.0
        ret1 = [0.0] + list(ret)
#         df['ret1'] = ret1
        ret1  = np.array(ret1)
#         ret2 = [0.0] + list((ret1[1:] + ret1[:-1]) / 2)
        df['ret2e'] = ema(ret1, 2)
        df['ret4e'] = ema(ret1, 4)
        df['ret8e'] = ema(ret1, 8)
        df['ret16e'] = ema(ret1, 16)
        df['ret32e'] = ema(ret1, 32)
        df['ret64e'] = ema(ret1, 64)
        df['ret128e'] = ema(ret1, 128)
        
    def add_std_ret(df):
        ''' add std of return calculated by closing price '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ret = cps[1:] / cps[:-1] - 1.0
        ret1 = [0.0] + list(ret)
        ret1  = np.array(ret1)
        df['std2'] = stds(ret1, 2)
        df['std4'] = stds(ret1, 4)
        df['std8'] = stds(ret1, 8)
        df['std16'] = stds(ret1, 16)
        df['std32'] = stds(ret1, 32)
        df['std64'] = stds(ret1, 64)
        df['std128'] = stds(ret1, 128)
        
        
    def add_ma_vol(df):
        ''' add ma of volume '''
        vol = df.loc[:, 'Volume'].values
        vol = np.array([float(val) for val in vol], dtype = 'float')
        df['vol2'] = ma(vol, 2)
        df['vol4'] = ma(vol, 4)
        df['vol8'] = ma(vol, 8)
        df['vol16'] = ma(vol, 16)
        df['vol32'] = ma(vol, 32)
        df['vol64'] = ma(vol, 64)
        df['vol128'] = ma(vol, 128)
        
    def add_vwap_c(df):
        ''' add difference between vwap and close'''
        df['vcb1'] = df['VWAP'].values / df['Close'].values - 1
        df['vcb2'] = ma(df['vcb1'].values, 2)
        df['vcb4'] = ma(df['vcb1'].values, 4)
        df['vcb8'] = ma(df['vcb1'].values, 8)
        df['vcb16'] = ma(df['vcb1'].values, 16)
        df['vcb32'] = ma(df['vcb1'].values, 32)
        df['vcb64'] = ma(df['vcb1'].values, 64)
        df['vcb128'] = ma(df['vcb1'].values, 128)
        
    def add_weekday_dummyvar(df):
        ''' add what week day it is '''
        dates = df['Date'].values
        dates = [datetime.date(int(val[:4]), int(val[5:7]), int(val[8:])) for val in dates]
        weekday = [(val.weekday() == 5 or val.weekday() == 6)  for val in dates]
        result = np.zeros(len(df))
        result[weekday] = 1
        df['weekday'] = result
        
    def add_subline(df):
        ''' add k-line subline '''
        length = len(df)
        result = np.ones(length)
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        result = np.min(np.vstack((ops, cps)), axis = 0)
        df['sub'] = (result - lows) / cps
        ops_2 = np.array([ops[0]] * 1 + list(ops[:-1]))
        df['sub2'] = (np.min(np.vstack((ops_2, cps)), axis = 0) - prev_lows(lows, 2)) / cps
        ops_4 = np.array([ops[0]] * 3 + list(ops[:-3]))
        df['sub4'] = (np.min(np.vstack((ops_4, cps)), axis = 0) - prev_lows(lows, 4)) / cps
        ops_8 = np.array([ops[0]] * 7 + list(ops[:-7]))
        df['sub8'] = (np.min(np.vstack((ops_8, cps)), axis = 0) - prev_lows(lows, 8)) / cps
        ops_16 = np.array([ops[0]] * 15 + list(ops[:-15]))
        df['sub16'] = (np.min(np.vstack((ops_16, cps)), axis = 0) - prev_lows(lows, 16)) / cps
        ops_32 = np.array([ops[0]] * 31 + list(ops[:-31]))
        df['sub32'] = (np.min(np.vstack((ops_32, cps)), axis = 0) - prev_lows(lows, 32)) / cps
        ops_64 = np.array([ops[0]] * 63 + list(ops[:-63]))
        df['sub64'] = (np.min(np.vstack((ops_64, cps)), axis = 0) - prev_lows(lows, 64)) / cps
        ops_128 = np.array([ops[0]] * 127 + list(ops[:-127]))
        df['sub128'] = (np.min(np.vstack((ops_128, cps)), axis = 0) - prev_lows(lows, 128)) / cps
        
    def add_macd(df):
        ''' add MACD '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        df['MACD'] = (ema(cps, 12) - ema(cps, 26)) / cps
        
    def add_bbbb_day(df):
        ''' add bbbb in day '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        highs = df.loc[:,'High'].values
        highs = np.array([float(val) for val in highs])
        df['bbbb_1'] = ((highs + lows) - 2.0 * cps) / cps
        df['bbbb_2'] = ((prev_highs(highs, 2) + prev_lows(lows, 2)) - 2.0 * cps) / cps
        df['bbbb_4'] = ((prev_highs(highs, 4) + prev_lows(lows, 4)) - 2.0 * cps) / cps
        df['bbbb_8'] = ((prev_highs(highs, 8) + prev_lows(lows, 8)) - 2.0 * cps) / cps
        df['bbbb_16'] = ((prev_highs(highs, 16) + prev_lows(lows, 16)) - 2.0 * cps) / cps
        df['bbbb_32'] = ((prev_highs(highs, 32) + prev_lows(lows, 32)) - 2.0 * cps) / cps
        df['bbbb_64'] = ((prev_highs(highs, 64) + prev_lows(lows, 64)) - 2.0 * cps) / cps
        df['bbbb_128'] = ((prev_highs(highs, 128) + prev_lows(lows, 128)) - 2.0 * cps) / cps
    
    def add_cccc_day(df):
        ''' add cccc in day '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        highs = df.loc[:,'High'].values
        highs = np.array([float(val) for val in highs])
        df['cccc_1'] = (highs - lows) / 2 / cps
        df['cccc_2'] = (prev_highs(highs, 2) - prev_lows(lows, 2)) / 2.0 / cps
        df['cccc_4'] = (prev_highs(highs, 4) - prev_lows(lows, 4)) / 2.0 / cps
        df['cccc_8'] = (prev_highs(highs, 8) - prev_lows(lows, 8)) / 2.0 / cps
        df['cccc_16'] = (prev_highs(highs, 16) - prev_lows(lows, 16)) / 2.0 / cps
        df['cccc_32'] = (prev_highs(highs, 32) - prev_lows(lows, 32)) / 2.0 / cps
        df['cccc_64'] = (prev_highs(highs, 64) - prev_lows(lows, 64)) / 2.0 / cps
        df['cccc_128'] = (prev_highs(highs, 128) - prev_lows(lows, 128)) / 2.0 / cps
        
    def add_kdj(df):
        ''' add kdj '''
        cps = df.loc[:,'Close'].values
        cps = np.array([float(val) for val in cps])
        ops = df.loc[:,'Open'].values
        ops = np.array([float(val) for val in ops])
        lows = df.loc[:,'Low'].values
        lows = np.array([float(val) for val in lows])
        highs = df.loc[:,'High'].values
        highs = np.array([float(val) for val in highs])
        
        # high freq
        K = np.zeros(len(df))
        for i in range(len(df)):
            start = max(i - 4, 0)
            end = i + 1
            K[i] = (cps[i] - np.min(lows[start:end])) / (np.max(highs[start:end]) - np.min(lows[start:end]))
        D = ema(K, 3)
        df['K'] = K
        df['D'] = D
        temp = np.ones(len(df))
        temp[K > 0.7] = 0
        temp[K < 0.3] = 0
        temp[K < D] = 0
        df['KDJ'] = temp
        
        # low freq shit
        K = np.zeros(len(df))
        for i in range(len(df)):
            start = max(i - 13, 0)
            end = i + 1
            K[i] = (cps[i] - np.min(lows[start:end])) / (np.max(highs[start:end]) - np.min(lows[start:end]))
        D = ema(K, 3)
        df['K_'] = K
        df['D_'] = D
        temp = np.ones(len(df))
        temp[K > 0.7] = 0
        temp[K < 0.3] = 0
        temp[K < D] = 0
        df['KDJ_'] = temp
        
    def standardize_ohlv(df):
        ''' divide open high low by vwapclose'''
        df['Open'] = df['Open'].values / df['Close'].values
        df['High'] = df['High'].values / df['Close'].values
        df['Low'] = df['Low'].values / df['Close'].values
        df['VWAP'] = df['VWAP'].values / df['Close'].values
        
    
    add_ma_ret(df)
    add_ema_ret(df)
    add_std_ret(df)
    add_ma_vol(df)
    add_weekday_dummyvar(df)
    add_subline(df)
    add_macd(df)
    add_bbbb_day(df)
    add_cccc_day(df)
    add_kdj(df)
    standardize_ohlv(df)
    return df
```
